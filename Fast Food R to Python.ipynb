```python
from sklearn.metrics import adjusted_rand_score
from sklearn.utils import resample
import os

np.random.seed(1234)
# Define the number of segments to evaluate
num_segments = range(2, 9)

# Define the number of bootstrap samples and random initializations
n_boot = 100
n_rep = 10
os.environ['OMP_NUM_THREADS'] = '1'
```

```python
# Initialize an array to store stability scores
stability_scores = np.zeros((len(num_segments), n_boot))

```

```python
# Perform stability-based analysis for each number of segments
for i, k in enumerate(num_segments):
    # Perform bootstrap samples
    for j in range(n_boot):
        bootstrap_sample = resample(MD_x)
        
        # Perform k-means clustering with random initializations
        kmeans1 = KMeans(n_clusters=k, n_init=n_rep).fit(bootstrap_sample)
        labels = kmeans1.labels_
        
        # Calculate adjusted Rand index as stability score
        stability_scores[i, j] = adjusted_rand_score(labels, kmeans1.predict(MD_x))
```

```python
plt.figure(figsize=(12, 8))
plt.boxplot(stability_scores.T)
plt.xlabel('Number of segments')
plt.ylabel('Adjusted Rand Index')
plt.xticks(range(1, len(num_segments) + 1), num_segments)
plt.title('Global Stability Boxplot')
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score

# Set the random seed for reproducibility
np.random.seed(1234)

# Function to calculate stability for a given number of segments
def calculate_stability(X, n_segments, nboot, nrep):
    stability_scores = []

    for _ in range(nboot):
        bootstrap_indices = np.random.choice(len(X), size=len(X), replace=True)
        bootstrap_X = X[bootstrap_indices]

        best_score = -1

        for _ in range(nrep):
            kmeans = KMeans(n_clusters=n_segments)
            labels = kmeans.fit_predict(bootstrap_X)
            score = adjusted_rand_score(labels, kmeans.labels_)
            if score > best_score:
                best_score = score

        stability_scores.append(best_score)

    return stability_scores

# Generate bootstrap samples for stability calculation
nboot = 100  # Number of bootstrap samples
nrep = 10  # Number of k-means initializations for each sample
n_segments = range(2, 9)  # Number of segments from 2 to 8

stability_scores = []

for n in n_segments:
    scores = calculate_stability(MD_x, n, nboot=nboot, nrep=nrep)
    stability_scores.append(scores)

# Generate the boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(stability_scores, labels=n_segments)
plt.xlabel('Number of Segments')
plt.ylabel('Adjusted Rand Index')
plt.title('Global Stability Boxplot')
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans

np.random.seed(1234)


# Clustering using KMeans with 2 to 8 clusters
k_range = range(2, 9)
n_rep = 10

cluster_results = []
for k in k_range:
    kmeans_clusters = []
    for _ in range(n_rep):
        kmeans = KMeans(n_clusters=k, random_state=1234)
        kmeans.fit(MD_x)
        kmeans_clusters.append(kmeans.labels_)
    cluster_results.append(kmeans_clusters)

# Relabeling the clusters for consistent interpretation
def relabel_clusters(cluster_results):
    relabeled_clusters = []
    for clusters in cluster_results:
        relabeled = []
        for labels in clusters:
            relabel_map = {old_label: new_label for new_label, old_label in enumerate(np.unique(labels))}
            relabeled.append([relabel_map[label] for label in labels])
        relabeled_clusters.append(relabeled)
    return relabeled_clusters

MD_km28 = relabel_clusters(cluster_results)
MD_k4 = MD_km28[3]
```

```python
# Calculate hierarchical cluster analysis on attributes
MD_vclust = linkage(np.transpose(MD_x), method='complete', metric='euclidean')

# Create the segment profile plot
fig, ax = plt.subplots(figsize=(8, 6))

# Define the attributes' names
attributes = ['Cheap', 'Greasy', 'Disgusting', 'Expensive', 'Tasty', 'Yummy', 'Healthy']

# Set the marker variables for each segment
marker_variables = {
    1: {'Cheap', 'Greasy'},
    2: {'Disgusting', 'Expensive'},
    3: {'Expensive', 'Tasty', 'Yummy'},
    4: {'Tasty', 'Yummy', 'Cheap', 'Healthy'}
}

# Define the overall sample percentages for each attribute
overall_percentages = [0.3, 0.2, 0.1, 0.4, 0.6, 0.5, 0.3]

# Iterate over each segment and plot the bars
for segment, marker_vars in marker_variables.items():
    # Get the indices of the marker variables
    marker_indices = [attributes.index(attr) for attr in marker_vars]
    # Get the percentages for the segment's attributes
    segment_percentages = np.mean(MD_k4[0] == segment, axis=0)
    # Get the differences between segment percentages and overall percentages
    differences = segment_percentages - overall_percentages
    # Plot the bars
    ax.bar(range(len(attributes)), segment_percentages, label=f'Segment {segment}')
    # Highlight the marker variables with a different color
    ax.bar(marker_indices, segment_percentages[marker_indices], color='red')
    # Add text labels for the differences greater than the thresholds
    for i, diff in enumerate(differences):
        if abs(diff) > 0.25 or abs(diff / overall_percentages[i]) > 0.5:
            ax.text(i, segment_percentages[i] + 0.02, f'{diff:.2f}', ha='center', color='black')

# Set the labels and title
ax.set_xticks(range(len(attributes)))
ax.set_xticklabels(attributes, rotation=45)
ax.set_xlabel('Attributes')
ax.set_ylabel('Percentage')
ax.set_title('Segment Profile Plot')

# Add a legend
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
